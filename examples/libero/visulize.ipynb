```python
import numpy as np
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("../../data/libero/logs.csv")

df.head()
```

```python
from plotly.subplots import make_subplots

task_id = 4
episode_idx = 0

# Filter the dataframe for the specific task and episode
plot_df = df[(df["task_id"] == task_id) & (df["episode_idx"] == episode_idx)]

# Subtract 80 cm (table offset) from state_eef_pos_z
# https://github.com/Lifelong-Robot-Learning/LIBERO/blob/8f1084e3132a39270c3a13ebe37270a43ece2a01/libero/libero/envs/arenas/kitchen_arena.py#L30C2-L32C34
plot_df["state_eef_pos_z"] -= 0.85


fig = make_subplots(rows=2, cols=1,
                    shared_xaxes=True)


# Visualize the state end-effector position coordinates over time
for trace in px.line(plot_df, x="step_idx", y=["state_eef_pos_x", "state_eef_pos_y", "state_eef_pos_z"]).data:
    fig.add_trace(trace, row=1, col=1)

# plot_df["action_norm"] = np.linalg.norm(plot_df[["action_0", "action_1", "action_2"]], axis=1)


# Visualize the action end-effector position coordinates over time
for trace in px.line(plot_df, x="step_idx", y=["action_0", "action_1", "action_2"]).data:
    fig.add_trace(trace, row=2, col=1)

fig.update_layout(title_text="State vs. Action EEF Position Space over Time")
fig.show()
```
