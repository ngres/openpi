```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patheffects as path_effects
from matplotlib.ticker import PercentFormatter

def finalize_paper_plot(ax):
    """
    Applies the custom 'paper' flair that .mplstyle can't handle.
    """
    # 1. Apply Shadow Effects to all bars/patches
    pe = [path_effects.withSimplePatchShadow(shadow_rgbFace="black", alpha=1)]
    for patch in ax.patches:
        patch.set_path_effects(pe)
        patch.set_edgecolor('black')
        patch.set_clip_on(False)

    # 3. Handle the grey labels/monospace consistently
    ax.xaxis.label.set_color('grey')
    ax.xaxis.label.set_family('monospace')

    # --- Usage Example ---
    plt.style.use('./academic_clean.mplstyle')

fig, ax = plt.subplots(figsize=(12, 4))

# Because of the prop_cycle, we don't even need to specify colors!
tasks = ['A', 'B', 'C', 'D']
vals1 = [70, 40, 50, 20]
vals2 = [30, 60, 20, 50]

x = np.arange(len(tasks))
ax.bar(x - 0.15, vals1, 0.2, label='Model A')
ax.bar(x + 0.15, vals2, 0.2, label='Model B')

ax.set_xticks(x)
ax.set_xticklabels(tasks)

ax.yaxis.set_major_formatter(PercentFormatter())

ax.set_ylabel('Success Rate')
ax.set_xlabel('50 real-world demonstrations per task')

# One call to rule them all
finalize_paper_plot(ax)

plt.show()
```
